## 产生-1···1间隔为0.05的向量 $ \vec x $

```python
import numpy as np
x=np.linspace(-1,1,41)
print("x的值为:\n",x)
```

## 计算方程$ y=x^2+3 $

```python
y=x*x+3
print("y的值为:\n",y)
```

## 绘制函数$ y=x^2+3 $的函数图像

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.show()
```

## 随机产生大小为30\*30的随机矩阵$ A $

```python
A=np.random.random((30,30))
print("矩阵A的值如下:\n",A)
```

## 绘制箱形图

```python
import pandas as pd
df=pd.DataFrame(A)
df.boxplot()
plt.ylabel('y')
plt.xlabel('x')
plt.title('$A$')
plt.show()
```

## 在矩阵$ A $中随机选择100个数据

```python
# 先将B转化成一维数组 B与A共享一块内存
B=A.reshape(900,)
import random
C=random.sample(range(0,901),100) # 从0-900中随机生成不重复的数字
BB=B[C] # BB即为从A中取出的100个数组成的array
print("随机选择的数据如下:\n",BB)
```

## 用100个2-8之间的随机数替换矩阵A中的数据

```python
D=[random.randint(2,8) for i in range(100)] # 生成范围在2-8之间的随机整数 100个
B[C]=D # 替换
print("替换后的矩阵A为:\n",A)
```
