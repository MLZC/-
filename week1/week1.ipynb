- 随机产生大小为30\*30的随机矩阵$ A $

```python
print("矩阵A的值如下:\n",A)
```

- 绘制箱形图

```python
A.shape
```

- 在矩阵$ A $中随机选择100个数据

```python
B=A.reshape(900,)
```

```python
B.shape # 先将B转化成一维数组 B与A共享一块内存
```

```python
import random
C=random.sample(range(0,901),100)
```

```python
C # 从0-900中随机生成不重复的数字
```

```python
BB=B[C] #从B中随机取100个数
```
